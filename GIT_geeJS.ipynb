## GitHub for GEE JavaScipt API 
Create a new github repository from https://github.com/new (Do NOT initialize with README / .gitignore)

Clone from googlesource (with authentication)

```python
git clone https://earthengine.googlesource.com/users/your_username/CCG
cd CCG
```

If prompted, use the generated identifier from:
https://earthengine.googlesource.com/new-password

Add GitHub as a remote

```python
git remote add github https://github.com/<your-username>/CCG.git
```

Verify

```python
git remote -v
```

Push everything to GitHub

```python
git push github --all
git push github --tags
```

Update Github repository

```python
git fetch origin && git push github
```
